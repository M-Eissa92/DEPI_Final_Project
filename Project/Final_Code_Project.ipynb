```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

# Load the dataset from the specified directory
file_path = "Retail-Supply-Chain-Sales-Dataset.xlsx"
df = pd.read_excel(file_path)

# Display initial data and info
display(df.head())
df.info()

# Convert 'Order Date' to datetime format and set it as the index
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.set_index('Order Date', inplace=True)

# === Step 1: Exploratory Data Analysis (EDA) ===

# Summary statistics
print("\nSummary Statistics:")
display(df.describe(include='all'))

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Distribution of Sales
plt.figure(figsize=(10, 6))
sns.histplot(df['Sales'], bins=50, kde=True, color="blue")
plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()

# Boxplot for Sales to detect outliers
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['Sales'])
plt.title("Boxplot of Sales")
plt.show()

# Correlation heatmap
columns_to_exclude = ['Row ID', 'Postal Code']     # Exclude non-numeric columns for correlation analysis
correlation_matrix = df.drop(columns=columns_to_exclude).corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# Bar chart of average sales by category
category_sales = df.groupby('Category')['Sales'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=category_sales.index, y=category_sales.values, palette="viridis")
plt.title("Average Sales by Category")
plt.xlabel("Category")
plt.ylabel("Average Sales")
plt.xticks(rotation=45)
plt.show()

# Resample data to monthly sales
monthly_sales = df['Sales'].resample('ME').sum()

# Plot monthly sales trend
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales, marker='o', linestyle='-')
plt.title("Monthly Sales Trend")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()

# Decompose the time series into trend, seasonality, and residuals
decomposition = seasonal_decompose(monthly_sales, model='additive', period=12)

# Plot decomposition
fig, axes = plt.subplots(4, 1, figsize=(12, 8), sharex=True)
axes[0].set_title("Time Series Decomposition")
axes[0].plot(monthly_sales, label="Original")
axes[1].plot(decomposition.trend, label="Trend")
axes[2].plot(decomposition.seasonal, label="Seasonality")
axes[3].plot(decomposition.resid, label="Residuals")
for ax in axes:
    ax.legend(loc="upper left")
plt.tight_layout()
plt.show()

# === Step 2: Handle Missing Values ===
# Fill missing numeric values with median
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# Check for duplicates
print("\nNumber of Duplicate Rows:", df.duplicated().sum())

# Remove duplicates
df.drop_duplicates(inplace=True)

# === Step 3: Handle Outliers ===
def handle_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    print(f"Outliers for {column}: Lower Bound = {lower_bound}, Upper Bound = {upper_bound}")
    
    # Cap outliers instead of removing them
    data[column] = np.where(data[column] < lower_bound, lower_bound, data[column])
    data[column] = np.where(data[column] > upper_bound, upper_bound, data[column])
    return data

# Apply IQR-based outlier handling to numeric columns
outlier_columns = ['Sales', 'Discount', 'Profit', 'Quantity']
for col in outlier_columns:
    df = handle_outliers_iqr(df, col)

# Revisit boxplots after handling outliers
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['Sales'])
plt.title("Boxplot of Sales (After Handling Outliers)")
plt.show()

# Log transformation for skewed columns like Sales and Profit
df['Log_Sales'] = np.log1p(df['Sales'])
df['Log_Profit'] = np.log1p(df['Profit'])

# Visualize the effect of log transformation
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['Sales'], bins=50, kde=True, color="blue")
plt.title("Distribution of Sales (Before Log Transformation)")

plt.subplot(1, 2, 2)
sns.histplot(df['Log_Sales'], bins=50, kde=True, color="green")
plt.title("Distribution of Sales (After Log Transformation)")
plt.tight_layout()
plt.show()

# === Step 4: Encode Categorical Columns ===
# Drop irrelevant columns (e.g., Order ID, Customer ID)
df.drop(columns=['Order ID', 'Customer ID'], inplace=True)

# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

# Exclude 'Returned' column (binary flag) from one-hot encoding
if 'Returned' in categorical_cols:
    categorical_cols.remove('Returned')

# Apply one-hot encoding to categorical columns
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Convert 'Returned' column to binary (0 = Not Returned, 1 = Returned)
df['Returned'] = df['Returned'].apply(lambda x: 1 if x == 'Yes' else 0)

# === Step 5: Feature Engineering ===
# Create time-based features
df['Year'] = df.index.year
df['Month'] = df.index.month
df['Day'] = df.index.day
df['Weekday'] = df.index.dayofweek

# Create multiple lag features
for lag in [1, 2, 3, 6, 12]:  # Include lags for 1, 2, 3, 6, and 12 months
    df[f'Lag_{lag}'] = df['Sales'].shift(lag)

# Create rolling statistics (e.g., rolling mean and rolling standard deviation)
df['Rolling_Mean_3'] = df['Sales'].rolling(window=3).mean()
df['Rolling_Mean_6'] = df['Sales'].rolling(window=6).mean()
df['Rolling_Std_3'] = df['Sales'].rolling(window=3).std()
df['Rolling_Std_6'] = df['Sales'].rolling(window=6).std()

# Drop rows with NaN values caused by lagging and rolling operations
df.dropna(inplace=True)

# === Step 6: Save Preprocessed Data ===
preprocessed_file_path = "Preprocessed_Retail_Sales_Data.csv"
df.to_csv(preprocessed_file_path, index=True)

print(f"\nPreprocessed data saved to {preprocessed_file_path}")

# Summarize key insights
print("\nKey Insights:")
print("- The dataset contains sales data from multiple regions and categories.")
print("- Strong correlations exist between features like 'Discount' and 'Sales'.")
print("- Seasonal patterns are evident in the monthly sales trend.")
print("- Outliers were detected and mitigated using capping and log transformation.")
print("- Time-based, lagged, and rolling features were created for modeling purposes.")
print("- Categorical variables were encoded using one-hot encoding.")
print("- The dataset is now clean and ready for advanced analysis and modeling.")
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.tseries.holiday import USFederalHolidayCalendar
from datetime import timedelta
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer

# Load the preprocessed dataset from Milestone 1
preprocessed_file_path = "Preprocessed_Retail_Sales_Data.csv"
df = pd.read_csv(preprocessed_file_path, parse_dates=['Order Date'])
df.set_index('Order Date', inplace=True)

# === Step 1: Integrate Holiday Data ===
cal = USFederalHolidayCalendar()
holidays = cal.holidays(start=df.index.min(), end=df.index.max())

# Add a binary flag for holidays
df['Is_Holiday'] = df.index.isin(holidays).astype(int)

# Add a rolling holiday effect (e.g., 7 days before and after a holiday)
df['Holiday_Week'] = df.index.to_series().apply(
    lambda x: 1 if any(x == holiday + timedelta(days=i) for holiday in holidays for i in range(-7, 8)) else 0
)

# Visualize holiday effects on sales
plt.figure(figsize=(12, 6))
sns.boxplot(x='Is_Holiday', y='Sales', data=df)
plt.title("Impact of Holidays on Sales")
plt.xlabel("Is Holiday (1 = Yes, 0 = No)")
plt.ylabel("Sales")
plt.show()

# Visualize rolling holiday effect on sales
plt.figure(figsize=(12, 6))
sns.boxplot(x='Holiday_Week', y='Sales', data=df)
plt.title("Impact of Rolling Holiday Window (±7 Days) on Sales")
plt.xlabel("In Holiday Week (1 = Yes, 0 = No)")
plt.ylabel("Sales")
plt.show()

# === Step 2: Integrate Weather Data (Simulated Example) ===
np.random.seed(42)  # For reproducibility
df['Temperature'] = np.random.uniform(low=30, high=100, size=len(df))  # Simulated temperature in Fahrenheit
df['Precipitation'] = np.random.uniform(low=0, high=5, size=len(df))   # Simulated precipitation in inches

# Visualize the relationship between weather and sales
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Temperature vs. Sales
sns.scatterplot(x='Temperature', y='Sales', data=df, ax=axes[0])
axes[0].set_title("Temperature vs. Sales")
axes[0].set_xlabel("Temperature (°F)")
axes[0].set_ylabel("Sales")

# Precipitation vs. Sales
sns.scatterplot(x='Precipitation', y='Sales', data=df, ax=axes[1])
axes[1].set_title("Precipitation vs. Sales")
axes[1].set_xlabel("Precipitation (inches)")
axes[1].set_ylabel("Sales")

plt.tight_layout()
plt.show()

# === Step 3: Advanced Temporal Features ===
df['Quarter'] = df.index.quarter
df['Week_of_Year'] = df.index.isocalendar().week
df['Is_Weekend'] = df.index.dayofweek >= 5  # Saturday = 5, Sunday = 6

# === Step 4: Cyclical Features ===
df['Month_Sin'] = np.sin(2 * np.pi * df.index.month / 12)
df['Month_Cos'] = np.cos(2 * np.pi * df.index.month / 12)
df['Day_Sin'] = np.sin(2 * np.pi * df.index.day / 31)
df['Day_Cos'] = np.cos(2 * np.pi * df.index.day / 31)

# Visualize cyclical features
fig, axes = plt.subplots(1, 2, figsize=(14, 6), subplot_kw={'projection': 'polar'})

# Month cyclical encoding
axes[0].scatter(df['Month_Cos'], df['Month_Sin'], c=df['Sales'], cmap='viridis', s=10)
axes[0].set_title("Cyclical Encoding of Month")
axes[0].set_xticks(np.linspace(0, 2 * np.pi, 12, endpoint=False))
axes[0].set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Day cyclical encoding
axes[1].scatter(df['Day_Cos'], df['Day_Sin'], c=df['Sales'], cmap='viridis', s=10)
axes[1].set_title("Cyclical Encoding of Day")
axes[1].set_xticks(np.linspace(0, 2 * np.pi, 31, endpoint=False))
axes[1].set_xticklabels([str(i) for i in range(1, 32)], fontsize=8)

plt.tight_layout()
plt.show()

# === Step 5: Economic Indicators (Simulated Example) ===
np.random.seed(42)  # For reproducibility
df['Inflation_Rate'] = np.random.uniform(low=1, high=5, size=len(df))  # Simulated inflation rate (%)
df['Unemployment_Rate'] = np.random.uniform(low=3, high=8, size=len(df))  # Simulated unemployment rate (%)

# Visualize economic indicators vs. sales
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Inflation Rate vs. Sales
sns.scatterplot(x='Inflation_Rate', y='Sales', data=df, ax=axes[0])
axes[0].set_title("Inflation Rate vs. Sales")
axes[0].set_xlabel("Inflation Rate (%)")
axes[0].set_ylabel("Sales")

# Unemployment Rate vs. Sales
sns.scatterplot(x='Unemployment_Rate', y='Sales', data=df, ax=axes[1])
axes[1].set_title("Unemployment Rate vs. Sales")
axes[1].set_xlabel("Unemployment Rate (%)")
axes[1].set_ylabel("Sales")

plt.tight_layout()
plt.show()

# === Step 6: Interaction Features ===
df['Quantity_Discount'] = df['Quantity'] * df['Discount']

# === Step 7: Rolling Statistics ===
# Rolling variance to capture volatility
df['Rolling_Variance_6'] = df['Sales'].rolling(window=6).var()

# Plot seasonal patterns using rolling mean
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Rolling_Mean_3'], label="Rolling Mean (3 Months)", color="blue")
plt.plot(df.index, df['Rolling_Mean_6'], label="Rolling Mean (6 Months)", color="orange")
plt.title("Seasonal Patterns in Sales")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()

# === Step 8: Dimensionality Reduction with PCA ===
# Handle Missing Values Before PCA ===
# Check for missing values
print("\nMissing Values Before Imputation:")
print(df.isnull().sum())

# Impute missing numeric values using the median
numeric_cols = df.select_dtypes(include=[np.number]).columns
imputer = SimpleImputer(strategy='median')
df[numeric_cols] = imputer.fit_transform(df[numeric_cols])

# Drop rows with any remaining NaN values (if necessary)
df.dropna(inplace=True)

# Verify that no missing values remain
print("\nMissing Values After Imputation:")
print(df.isnull().sum())

# Standardize numeric features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[numeric_cols])

# Apply PCA
pca = PCA(n_components=0.95)  # Retain enough components to explain 95% of the variance
pca_results = pca.fit_transform(df_scaled)
print(f"Number of Components: {pca.n_components_}")
print(f"Explained Variance Ratio: {pca.explained_variance_ratio_.sum()}")

# Create a DataFrame for PCA results
# Create a DataFrame for PCA results
pca_columns = [f'PC{i+1}' for i in range(pca_results.shape[1])]  # Dynamically generate column names
pca_df = pd.DataFrame(data=pca_results, columns=pca_columns, index=df.index)

# Plot PCA results
plt.figure(figsize=(10, 6))
plt.scatter(pca_df['PC1'], pca_df['PC2'], alpha=0.5)
plt.title("PCA Results")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.show()

# Print explained variance ratio
print("\nExplained Variance Ratio:")
print(pca.explained_variance_ratio_)

# === Step 9: Drop Rows with NaN Values ===
# Drop rows with NaN values caused by rolling operations
df.dropna(inplace=True)

# === Step 10: Save Advanced Preprocessed Data ===
advanced_preprocessed_file_path = "Advanced_Preprocessed_Retail_Sales_Data.csv"
df.to_csv(advanced_preprocessed_file_path, index=True)

print(f"\nAdvanced preprocessed data saved to {advanced_preprocessed_file_path}")

# === Step 11: Visualize Correlations for New Features ===
# Compute correlation matrix for numeric features
# Subset features for correlation analysis
key_features = ['Sales', 'Profit', 'Discount', 'Quantity']
new_features = [
    'Is_Holiday', 'Holiday_Week', 'Temperature', 'Precipitation',
    'Quarter', 'Week_of_Year', 'Is_Weekend',
    'Inflation_Rate', 'Unemployment_Rate',
    'Quantity_Discount', 'Rolling_Variance_6','Lag_1', 'Lag_3', 'Lag_6', 'Lag_12',
    'Rolling_Mean_3', 'Rolling_Mean_6',
]
features_to_analyze = key_features + new_features

# Subset the DataFrame
df_subset = df[features_to_analyze]

# Compute correlation matrix
correlation_matrix = df_subset.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap for Key Features and New Features")
plt.show()

# Summarize key insights
print("\nKey Insights:")
print("- Holiday flags were added to capture the impact of holidays on sales.")
print("- Simulated weather data (temperature and precipitation) was integrated.")
print("- Advanced temporal features like quarter, week of the year, and weekend flags were created.")
print("- Rolling statistics (mean, standard deviation, and variance) were computed to smooth trends.")
print("- Cyclical features (month and day) were encoded using sine and cosine transformations.")
print("- Simulated economic indicators (inflation and unemployment rates) were added.")
print("- Interaction features (e.g., Quantity * Discount) were created.")
print("- Dimensionality reduction with PCA identified the most important features.")
print("- Correlations between new features and key variables (Sales, Profit, Discount, Quantity) were visualized.")
print("- The dataset is now enriched with external and advanced features for modeling.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns   
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
import matplotlib.pyplot as plt
import joblib

# === Step 1: Load Preprocessed Data ===
file_path = "Advanced_Preprocessed_Retail_Sales_Data.csv"
df = pd.read_csv(file_path, parse_dates=['Order Date'])
df.set_index('Order Date', inplace=True)

#df['Order Date'] = pd.to_datetime(df['Order Date'])
#df = df.set_index('Order Date')

df['Quantity_Lag1'] = df['Quantity'].shift(1)
df['Quantity_Lag4'] = df['Quantity'].shift(4)
df['Quantity_Lag8'] = df['Quantity'].shift(8)
df['Quantity_Lag12'] = df['Quantity'].shift(12)
df['Quantity_Lag52'] = df['Quantity'].shift(52)
df = df.dropna()

weekly_data = df[['Quantity', 'Quantity_Lag1', 'Quantity_Lag4', 'Quantity_Lag8','Quantity_Lag12', 'Quantity_Lag52' ]].resample('W').sum()
X = weekly_data[['Quantity_Lag1', 'Quantity_Lag4', 'Quantity_Lag8','Quantity_Lag12', 'Quantity_Lag52']]
y = weekly_data['Quantity']

X = X.reindex(y.index)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# ------------------------------------------
# 🤖 Step 3A: Random Forest Model
# ------------------------------------------
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)

# Evaluation
rf_metrics = {
    'MAE': mean_absolute_error(y_test, y_pred_rf),
    'MSE': mean_squared_error(y_test, y_pred_rf),
    'RMSE': np.sqrt(mean_squared_error(y_test, y_pred_rf)),
    'R2': r2_score(y_test, y_pred_rf)
}

# ------------------------------------------
# 🤖 Step 3B: XGBoost Model
# ------------------------------------------
xgb_model = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=3)
xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.predict(X_test)

xgb_metrics = {
    'MAE': mean_absolute_error(y_test, y_pred_xgb),
    'MSE': mean_squared_error(y_test, y_pred_xgb),
    'RMSE': np.sqrt(mean_squared_error(y_test, y_pred_xgb)),
    'R2': r2_score(y_test, y_pred_xgb)
}

# ------------------------------------------
# 📊 Step 4: Comparison
# ------------------------------------------
print("\n📈 Random Forest Performance:")
for k, v in rf_metrics.items():
    print(f"{k}: {v:.2f}")

print("\n📈 XGBoost Performance:")
for k, v in xgb_metrics.items():
    print(f"{k}: {v:.2f}")

# Visualize Predictions vs Actual
plt.figure(figsize=(12, 5))
plt.plot(y_test.index, y_test.values, label='Actual', marker='o')
plt.plot(y_test.index, y_pred_rf, label='RF Predicted', linestyle='--')
plt.plot(y_test.index, y_pred_xgb, label='XGB Predicted', linestyle=':')
plt.legend()
plt.title('Forecasting Sales - Actual vs Predicted')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()

# ------------------------------------------
# 💾 Step 5: Save Best Model
# ------------------------------------------
best_model = xgb_model if xgb_metrics['RMSE'] < rf_metrics['RMSE'] else rf_model
joblib.dump(best_model, 'best_forecasting_model_weekly.pkl')
print("✅ Best model saved as 'best_forecasting_model_weekly.pkl'")
joblib.dump(X_train.columns.tolist(), 'training_columns_weekly.pkl')
```
